### Group By

```python
# Group by will split the object based on the column we specify
spec = iris.groupby("species")
spec.head(2)
```

```python
spec
```

```python
# We need to iterate through this object
for species in spec:
    print(species)
```

```python
# We can still use DF functions on this object
spec.describe()
```
