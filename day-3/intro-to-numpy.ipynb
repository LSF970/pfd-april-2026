# Intro to NumPy

NumPy is a library which provides support for large, homogenous, multi-dimensional arrays and matrices.

```python
# Importing NumPy library
import numpy as np
```

## Creating Arrays in NumPy

**Creating a NumPy array from a Python list**

```python
list1 = [10, 20, 30]
list1
```

```python
arr1 = np.array(list1)
arr1
```

```python
type(arr1)
```

**Using arange to create an array**

The difference between .array and .arange is that .array takes a list and turns it into an array. Whereas .arange creates an array from a range of numbers.

```python
arr2 = np.arange(10, 20, 1) # (Start, Stop (inclusive), Step-Size) 
arr2
```

## Multidimensional arrays

**Combining two lists --> Two dimensional arrays**

When we combine two arrays together, we add dimensions. So combining two arrays creates a 2d array.

This then allows us to perform matrix operations --> This is why we use NumPy.

```python
# Create a new list to add to the first list
list2 = [40, 50, 60]

# Join the lists together
list3 = [list1, list2]
```

```python
# Create an array from the joined lists and output, notice the difference?
arr3 = np.array(list3)
arr3
```

```python
# Let's find the shape of the array
# We have a 2d array, 2 elementsin the first instance (dimension) and 3 in the second (dimension)
arr3.shape
```

**Three dimensional arrays**

```python
arr3dim = np.array([[[1,2,3],[3,4,5]],[[5,6,7],[7,8,9]]])
arr3dim
```

```python
arr3dim.shape # 2 sets of list, set lists each set, 3 element per list --> 2, 2, 3
```

## Using Arrays and Scalars

NumPy arrays allows us to do matrix operations, see below:

```python
# For example, simple multiplication
arr3dim * 3
```

```python
# How is this different to doing:
list3 * 3
```

Python lists apply the multiplication to the lists themselves, NumPy arrays apply the multiplication to the contents!

**Multiply an array by another array, using a matrix operation:**

```python
# Creating another array which has the same size as arr1
arr11 = np.array(list2)

# Individual numbers will be multiplied based on their positions
# 10 * 40, 20 * 50, 30 * 60 
arr1 * arr11
```

```python
# What if we tried to multiply arrays of different shapes?
arr1 * arr2
```

```python
# You can however, multiply together arrays of different dimension but the same element size:
arr1 * arr3dim
```

## Accessing values in NumPy arrays

```python
# Can we access the 60 value from this array?
arr3
```

```python
# We can use indexing here
arr3[1][2]
```

```python
# Even faster way to do it
arr3[1, 2]
```

```python
# Access the "bottom" row of the array
arr3[1]
```

```python
# Access each row (:) and then output only the 3rd index (2)
arr3[:,2]
```

## Statistical functions and arrays

There are a number of functions built into NumPy we can use to get things like sum, averages etc.

```python
# Finding the sum of all values in an array
print(arr3.sum())
```

```python
# Finding the sum of values column wise
print(arr3.sum(0))
```

```python
# Finding the sum of values row wise
print(arr3.sum(1))
```

```python
# Finding the mean
print(arr3.mean())
```

```python
# Finding standard deviation
print(round(arr3.std(),2))
```

```python
# Finding variance
print(arr3.var())
```

```python
# min and max
print(arr3.min())

print(arr3.max())
```

## Iterating through NumPy arrays

```python
for x in arr1:
    print(x)
```

```python
# Looping through n-dimensional arrays --> nest your loops
for dim in arr3:
    for num in dim:
        print(num)
```

```python
# nditer() stands for n-dimensional iteration. Neater for nested loop when working with arrays.
for x in np.nditer(arr3):
    print(x)
```

## Saving and loading in NumPy

```python
# Save an array a file
np.save("new", arr3)
```

```python
# Load the file
loaded_arr = np.load("new.npy")
loaded_arr
```

```python
# Saving as .txt file
np.savetxt("new.txt", arr3, delimiter=",")
```

```python
# Loading a .txt file
loaded_from_txt = np.loadtxt("new.txt", delimiter=",")
loaded_from_txt
```
