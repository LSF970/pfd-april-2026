# Intro to Pandas

Pandas is a high-level data manipulation package/library which was built on top of NumPy. The key structures within pandas include Series and DataFrames.

```python
# Importing NumPy and Pandas
import numpy as np
import pandas as pd
# from pandas import Series, DataFrame
```

```python
# Creating a Series out of a list --> A Series is a 1 dimensional array with labels
x_series = pd.Series([10,20,30,40,50])
x_series
```

```python
# We can access different components of the Series if we want

# Index
print(x_series.index)

# Values
print(x_series.values)

# DType
print(x_series.dtype)
```

```python
# Create a Series with an index
data = [450, 650, 870]
Sales = pd.Series(data, index=["Don", "Mike", "Edwin"])
Sales
```

```python
# check the type
print(type(Sales))
```

```python
# Accesssing Values

# Using index name
print(Sales["Don"])

# Using indexing
print(Sales.iloc[0])
```

### Checking for conditions

```python
# We can filter our data based on conditions that we specify. We usually do this with booleans.

# Sales greater than 500
Sales>500
```

```python
# Return the values greater than 500
Sales[Sales>500]
```

```python
# Check for index names in a Series
"Don" in Sales
```

```python
# We can only check if something is in an index. See below:
450 in Sales
```

### Working with Dictionaries

```python
# Converting a Series to a Dictionary
sales_dict = Sales.to_dict()
sales_dict
```

```python
# Converting a Dictionary to a Series
Sales_ser = pd.Series(sales_dict)
Sales_ser
```

### Adding entries and working with NaN/Null values

```python
# We can create a new Series from an existing Series
# If we specify names in the index that were not there already, they will be assigned NaN values
# NaN = Not a Number

New_sales = pd.Series(Sales, index=["Don", "Mike", "Sally", "Edwin", "Lucy"])
New_sales
```

```python
# Check if entries/value are NaN --> Can use NumPy for this
print(np.isnan(New_sales["Sally"]))
```

```python
# This is different to None
New_sales["Sally"] is None
```

```python
# We can check for NaN's on the entire Series
np.isnan(New_sales)
```

```python
# Checking for null values with Pandas
pd.isnull(New_sales)
```

### Naming components in a Series

```python
# Naming an index
Sales.index.name = "SalesPerson"
Sales
```

```python
# Naming a Series
Sales.name = "Total TV Sales"
Sales
```

## DataFrames

DataFrames are two-dimensional, size-mutable, potentially heterogenous(diverse) tabular data structures.

Essentially a data structure with labelled axes (rows and columns).

### Creating DataFrames

```python
# Creating a DF from a list
data = [["Adrian", 20],["Bethany", 23],["Chloe", 41]]

# Remember to add the column names if making a DF from a list
df1 = pd.DataFrame(data, columns=["Name", "Age"])
df1
```

```python
# Creating a DF from a Dictionary
new_dict = {'Name':['Tom','Jane','Steve','Lucy'],'Sales':[250,300,350,400]}

df_dict1 = pd.DataFrame(new_dict)
df_dict1
```

```python
# Adding a custom index
df_dict_index1 = pd.DataFrame(new_dict, index=["rank4", "rank3", "rank2", "rank1"])
df_dict_index1
```

```python
# Creating a DF from a list of Dictionaries
# Why useful? --> Mimics JSON data
dict_list = [{'Name':'Tom','Sales':250},{'Name':'Jane','Sales':300},{'Name':'Steve','Sales':350}
            ,{'Name':'Lucy','Sales':400}]

df_dict_list = pd.DataFrame(dict_list)
df_dict_list
```

```python
# Combining Pandas Series into a single DataFrame part 1 - Creating the Series
east = pd.Series([1000,1200,3400],index=['Q1','Q2','Q3'])
west = pd.Series([1100,1300,2400,3500],index=['Q1','Q2','Q3','Q4'])
print(east)
print(west)
```

```python
# Combining Pandas Series into a single DataFrame part 2 - Creating the DF
df_regions = pd.DataFrame({"East":east, "West":west})
df_regions
```

```python
# Once we have a DataFrame, we can easily add Series on
df_regions["North"] = [2000,3000,2500,4000]
df_regions["South"] = [1500,2000,1500,4000]
df_regions
```

```python
# If we made a mistake and need to set a new index, we can add a new column
# and set that new column as the index

years = ['2020','2021','2022','2023']
df_regions["Years"] = years
df_regions
```

```python
# We can use set_index to set the index to a column in the DF
df_regions = df_regions.set_index("Years")
df_regions
```

```python
# Changing the index values, we can use reindex()
new_regions_df = df_regions.reindex(["2022", "2023", "2024", "2025", "2026"])
new_regions_df
```

```python
# reindexing columns
reindex_df = new_regions_df.reindex(columns=["North", "East", "South", "New"])
reindex_df
```
