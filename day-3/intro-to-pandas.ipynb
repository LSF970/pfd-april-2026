# Intro to Pandas

Pandas is a high-level data manipulation package/library which was built on top of NumPy. The key structures within pandas include Series and DataFrames.

```python
# Importing NumPy and Pandas
import numpy as np
import pandas as pd
from pandas import Series, DataFrame
```

```python
# Creating a Series out of a list --> A Series is a 1 dimensional array with labels
x_series = pd.Series([10,20,30,40,50])
x_series
```

```python
# We can acces different components of the Series if we want

# Index
print(x_series.index)

# Values
print(x_series.values)

# DType
print(x_series.dtype)
```

```python
# Create a Series with an index
data = [450, 650, 870]
Sales = pd.Series(data, index=["Don", "Mike", "Edwin"])
Sales
```

```python
# check the type
print(type(Sales))
```

```python
# Accesssing Values

# Using index name
print(Sales["Don"])

# Using indexing
print(Sales.iloc[0])
```

### Checking for conditions

```python
# We can filter our data based on conditions that we specify. We usually do this with booleans.

# Sales greater than 500
Sales>500
```

```python
# Return the values greater than 500
Sales[Sales>500]
```

```python
# Check for index names in a Series
"Don" in Sales
```

```python
# We can only check if something is in an index. See below:
450 in Sales
```
